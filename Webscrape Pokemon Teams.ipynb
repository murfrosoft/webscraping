Goal: Webscrape pokemon names, stats, and images from the website <a href="https://pokemondb.net/pokedex/national">pokemondb.net</a> and attempt to build two teams random teams of X number of pokemon each.

```python
import requests
from bs4 import BeautifulSoup
from time import sleep
import random
from IPython.display import display, Markdown
```

```python
def getSoup( url ):
    response = requests.get( url )
    soup = BeautifulSoup(response.content, 'html5lib')
    return soup
```

```python
def getStats( sub_url ):
    url = "https://pokemondb.net" + sub_url
    soup = getSoup(url)
    stat_table = soup.find_all('table', {'class':'vitals-table'})[3]
    stats = []
    trs = stat_table.find_all('td',{'class':'cell-num'})
    for i in range(len(trs)):
        if i%3 == 0:
            stats.append(int(trs[i].text))
    return tuple(stats)
```

```python
url = "https://pokemondb.net/pokedex/national"
soup = getSoup(url)
```

```python
pokemon = []
tags = soup.find_all('div', {'class':'infocard'})
number = 0
for div in tags:
    number = int(div.small.text[1:])
    name = div.small.parent.a.text
    sprite = div.span.a.span['data-src']
    types = div.find('a',{'class':'itype'}).parent.text
    link = div.a['href'] 
    stats = getStats(link)
    sleep(4)
    pokemon.append( (name, number, sprite, types) + stats)
    print(number, name, stats)
```

```python
team_size = 6
random.shuffle(pokemon)
markdown = '<h1>Team A</h1><br/><table><tr>'
for i in range(team_size):
    markdown += '<td><img src="%s"<br/>%s<br/># %d<br/>%s<br/>Power %d</td>' %(pokemon[i][2],pokemon[i][0],pokemon[i][1],pokemon[i][3],sum(pokemon[i][4:]))
markdown += '</tr></table>'
random.shuffle(pokemon)
markdown += '<br/><h1>Team B</h1><br/><table><tr>'
for i in range(team_size):
    markdown += '<td><img src="%s"<br/>%s<br/># %d<br/>%s<br/>Power %d</td>' %(pokemon[i][2],pokemon[i][0],pokemon[i][1],pokemon[i][3],sum(pokemon[i][4:]))
markdown += '</tr></table>'
display(Markdown(markdown))
```
